```python
import pandas as pd
import matplotlib.pyplot as plt
from cycler import cycler
```

```python
#import all TPA standard files
TPA_standards = []
for file in ['blank.txt','2.5 TPA.txt','5 TPA.txt','10 TPA.txt', '25 TPA.txt']:
    TPA_standards.append(pd.read_csv('Data/HPLC Data/' + file,sep='\t',header=None))
    
#adjust the blank trace to the same max peak height as the other traces
scale_factor = max(TPA_standards[0][1]) / max(TPA_standards[1][1])
TPA_standards[0][1] = TPA_standards[0][1] / scale_factor 
```

```python
#display full spectra

fig,ax = plt.subplots()
ax.plot(TPA_standards[0][0],TPA_standards[0][1])
ax.plot(TPA_standards[1][0],TPA_standards[1][1])
ax.plot(TPA_standards[2][0],TPA_standards[2][1])
ax.plot(TPA_standards[3][0],TPA_standards[3][1])
plt.show()
```

```python
#home in on TPA peak
TPA_min = 12.9
TPA_max = 13.5
fig,ax = plt.subplots()
ax.set_prop_cycle(cycler(color=['#cccccc','#888888','#444444','#000000']))
ax.plot(TPA_standards[0][TPA_standards[0][0].between(TPA_min,TPA_max)][0],TPA_standards[0][TPA_standards[0][0].between(TPA_min,TPA_max)][1], label='0 uM')
ax.plot(TPA_standards[1][TPA_standards[1][0].between(TPA_min,TPA_max)][0],TPA_standards[1][TPA_standards[1][0].between(TPA_min,TPA_max)][1], label='2.5 uM')
ax.plot(TPA_standards[2][TPA_standards[2][0].between(TPA_min,TPA_max)][0],TPA_standards[2][TPA_standards[2][0].between(TPA_min,TPA_max)][1], label='5 uM')
ax.plot(TPA_standards[3][TPA_standards[3][0].between(TPA_min,TPA_max)][0],TPA_standards[3][TPA_standards[3][0].between(TPA_min,TPA_max)][1], label='10 uM')
ax.set_ylabel('Absorption')
ax.set_xlabel('Time (sec)')
ax.set_ylim((0,22))
plt.legend(loc='upper left')
#plt.savefig('TPA_standard.svg')
plt.show()
```

```python
#import all BHET standard files
BHET_standards = []
for file in ['blank.txt','2.5 BHET.txt','5 BHET.txt','10 BHET.txt']:
    BHET_standards.append(pd.read_csv('Data/HPLC Data/' + file,sep='\t',header=None))
#adjust the blank trace to the same max peak height as the other traces
scale_factor = max(BHET_standards[0][1]) / max(BHET_standards[1][1])
BHET_standards[0][1] = BHET_standards[0][1] / scale_factor
```

```python
#display full spectra

fig,ax = plt.subplots()
ax.plot(BHET_standards[0][0],BHET_standards[0][1])
ax.plot(BHET_standards[1][0],BHET_standards[1][1])
ax.plot(BHET_standards[2][0],BHET_standards[2][1])
ax.plot(BHET_standards[3][0],BHET_standards[3][1])
plt.show()
```

```python
#home in on BHET peak
BHET_min = 11.5
BHET_max = 12.0
fig,ax = plt.subplots()
ax.set_prop_cycle(cycler(color=['#cccccc','#888888','#444444','#000000']))
ax.plot(BHET_standards[0][BHET_standards[0][0].between(BHET_min,BHET_max)][0],BHET_standards[0][BHET_standards[0][0].between(BHET_min,BHET_max)][1], label='0 uM')
ax.plot(BHET_standards[1][BHET_standards[1][0].between(BHET_min,BHET_max)][0],BHET_standards[1][BHET_standards[1][0].between(BHET_min,BHET_max)][1], label='2.5 uM')
ax.plot(BHET_standards[2][BHET_standards[2][0].between(BHET_min,BHET_max)][0],BHET_standards[2][BHET_standards[2][0].between(BHET_min,BHET_max)][1], label='5 uM')
ax.plot(BHET_standards[3][BHET_standards[3][0].between(BHET_min,BHET_max)][0],BHET_standards[3][BHET_standards[3][0].between(BHET_min,BHET_max)][1], label='10 uM')
ax.set_ylabel('Absorption')
ax.set_xlabel('Time (sec)')
ax.set_ylim((0,22))
plt.legend(loc='upper left')
#plt.savefig('BHET_standard.svg')
plt.show()
```
