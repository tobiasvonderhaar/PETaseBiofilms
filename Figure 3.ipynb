```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
CV_data = pd.read_csv("Data/170123 Crystal Violet Assay.csv")
```

```python
f, ax = plt.subplots()
sns.stripplot(data=CV_data, x = "Sample", y = "oD", palette='Greys')
#plt.savefig('Crystal Violet.svg')
plt.show()
```

```python
size_data = pd.read_csv("Data/230426 colony_size_smooth_rough.csv")
#apply colony size threshold 
area_size_threshold = 250
size_data_clean = size_data.drop(size_data[size_data['Area (µm2)'] > area_size_threshold].index)
print('Discarded ' + str(size_data.shape[0] - size_data_clean.shape[0]) + ' of ' + str(size_data.shape[0]) + ' datapoints.')
```

```python
f, ax = plt.subplots()
ax.set_yscale("log")
sns.boxplot(data=size_data_clean, x = "Sample", y = "Area (µm2)", width=0.6, palette = ['#c1efc9','#f9d3d8'], boxprops=dict(linewidth=1, edgecolor='k', alpha=1),fliersize = 0, notch = True)
sns.stripplot(data=size_data_clean, x = "Sample", y = "Area (µm2)", palette=['#228535','#ef7584'],alpha = 0.3)
ax.set_ylabel('Colony Area ($µm^2$)')
ax.set_xlabel('Region')
#plt.savefig('Colony_size.svg')
plt.show()
```
